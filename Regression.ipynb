```python
import pandas as pd
```

```python
data=pd.read_excel(r"Concrete_Data.xls")
data.head()
```

```python
data.columns
```

**Objective:** To come up with ML model that can predict Concrete compressive Strenght

```python
data.dtypes
```

```python
from sklearn.model_selection import train_test_split
### target and the predictors
ml_data=data.copy()
X=ml_data.drop(columns=["Concrete compressive strength(MPa, megapascals) "])
y=ml_data["Concrete compressive strength(MPa, megapascals) "]

#training set and testing set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)

print(f"training X:{X_train.shape} Y:{y_train.shape}")
print(f"training X:{X_test.shape} Y:{y_test.shape}")
```

```python
#Training a Linear regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_predictions=lr.predict(X_test)
```

**Model Evaluation**

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

r2=r2_score(y_test,y_predictions)
mse=mean_squared_error(y_test,y_predictions)
mae=mean_absolute_error(y_test,y_predictions)

print(f"R^2 : {r2} \n mse: {mse} \n mae: {mae}")
```

```python
#Training a Decision Tree regression model
from sklearn.tree import DecisionTreeRegressor
lr=DecisionTreeRegressor()
lr.fit(X_train,y_train)
y_predictions=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

r2=r2_score(y_test,y_predictions)
mse=mean_squared_error(y_test,y_predictions)
mae=mean_absolute_error(y_test,y_predictions)

print(f"R^2 : {r2} \n mse: {mse} \n mae: {mae}")
```
