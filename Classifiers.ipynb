```python
import pandas as pd
```

```python
train_data=pd.read_csv("train.csv")
train_data.head()
```



1.   Removing Items that may not have an impact on our model

*   PassengerId
*   Name


*   Ticket
*   Cabin number



```python
train_data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
train_data.head()
```

```python
train_data.dtypes
```

```python
train_data=pd.get_dummies(train_data,drop_first=True, dtype = int)
```

```python
train_data
```

```python
train_data.isnull().sum()
```

```python
train_data['Age']=train_data['Age'].fillna(train_data['Age'].median())
```

```python
train_data.isnull().sum()
```

**Training The model**

```python
from sklearn.model_selection import train_test_split
### target and the predictors
ml_data=train_data.copy()
X=ml_data.drop(columns=["Survived"])
y=ml_data["Survived"]

#training set and testing set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)

print(f"training X:{X_train.shape} Y:{y_train.shape}")
print(f"training X:{X_test.shape} Y:{y_test.shape}")
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)

y_prediction=dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

accuracy=accuracy_score(y_test,y_prediction)
confusion_matrix=confusion_matrix(y_test,y_prediction)

print(f"Accuracy:{accuracy}")
print(f"Confusion Matrix:{confusion_matrix}")
```
